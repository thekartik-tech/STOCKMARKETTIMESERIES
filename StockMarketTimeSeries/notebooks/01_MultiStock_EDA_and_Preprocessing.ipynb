```python

# Multi-Stock EDA and Preprocessing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA']

for symbol in symbols:
    filepath = f'../data/raw/{symbol}.csv'
    data = pd.read_csv(filepath, parse_dates=['Date'])
    data.set_index('Date', inplace=True)
    data = data.fillna(method='ffill')

    # Plot Close Price
    plt.figure(figsize=(10, 4))
    plt.plot(data['Close'], label=f'{symbol} Close Price')
    plt.title(f"{symbol} Closing Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()
    
    # Save processed data
    processed_path = f'../data/processed/{symbol}_cleaned.csv'
    data.to_csv(processed_path)
```
