```python

# Multi-Stock SARIMA Model

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA']

for symbol in symbols:
    print(f"Processing {symbol}...")
    data = pd.read_csv(f'../data/processed/{symbol}_cleaned.csv', parse_dates=['Date'])
    data.set_index('Date', inplace=True)

    # Fit SARIMA model (example seasonal order)
    model = SARIMAX(data['Close'], order=(1,1,1), seasonal_order=(1,1,1,12))
    sarima_fit = model.fit()

    # Forecast next 30 days
    forecast = sarima_fit.forecast(steps=30)

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(data['Close'], label='Historical')
    plt.plot(pd.date_range(data.index[-1], periods=30, freq='B'), forecast, label='Forecast')
    plt.title(f'{symbol} SARIMA Forecast')
    plt.legend()
    plt.show()
```
